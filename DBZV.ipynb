```python
dat1 = radar.azimuth['data']
dat2 = radar.elevation['data']
dat2 = dat2 * 4
```

```python
f = plt.figure(figsize=[15,8])
plt.plot(radar.time['data'], dat1 )
plt.plot(radar.time['data'], dat2 )
plt.xlabel(radar.time['standard_name'] + ' (' + radar.time['units'] + ')')
plt.ylabel(radar.azimuth['standard_name'] + ' (' + radar.azimuth['units'] + ')')
plt.legend(['Azimuth','Elevation'])
```
